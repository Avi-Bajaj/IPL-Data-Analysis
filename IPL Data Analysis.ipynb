# Q1 How many matches has each team has won (including home and away)? 

```python
# Merging file Match and Team 
a11 = pd.merge(match,team,left_on='Match_Winner_Id',right_on='Team_Id') 
a11.head(2)
```

```python
# Selecting Home team matches won
a12= a11.groupby(['Team_Name_Id'])['Match_Id'].count().sort_values(ascending=False)
a12
```

```python
# Selecting away team matches won 
a13= a11.groupby('Opponent_Team_Id')['Team_Name'].count().sort_values(ascending=False)
a13
```

```python
# Adding both the above results
a14=a13+a12
a14
```

```python
# Converting it to Data Frame
a15=pd.DataFrame(a14)
```

```python
# Reseting Index
a15.reset_index(inplace=True)
```

```python
# Renaming Columns Names
a16 = a15.rename(columns={'index':'Team_Id',0:'Matches_won'})
a16
```

```python
# Merging it with Team to get Team Names
a17 = pd.merge(team,a16,on='Team_Id')
```

```python
# Showing Results
a17
```

# Q2 Top 10 Run Getters IPL - ? 

## i.) Overall 

```python
# Merging files ball_by_ball and player for Batsman names
a21=pd.merge(ball_by_ball,player,left_on='Striker_Id',right_on='Player_Id')
a21.head()
```

```python
# Calculating top 10 runs scored
pd.DataFrame(a21.groupby('Player_Name')['Batsman_Scored'].sum().sort_values(ascending=False).head(10))
```

# ii.) Team Wise

```python
# Merging above created merged dataframe with team for team names 
a22 = pd.merge(a21,team,left_on='Team_Batting_Id',right_on='Team_Id')
```

```python
# Grouping dataframe by team name and player name 
a23 = a22.groupby(['Team_Name','Player_Name'])['Player_Name','Batsman_Scored'].sum().sort_values(by=['Team_Name','Batsman_Scored'],ascending=False)
a23.head()
```

```python
# resetting index to get top n (10) scorers
a23.reset_index(inplace=True)
```

```python
# setting index as player_name 
a23.set_index('Player_Name',inplace = True)
```

```python
# getting top n (top 5) scorers in each team
pd.DataFrame(a23.groupby('Team_Name')['Batsman_Scored'].nlargest(5))
```

# Q3 Who has scored most centuries?  

```python
# Merging files ball_by_ball and player to get batsman names
a31=pd.merge(ball_by_ball,player,left_on='Striker_Id',right_on='Player_Id')
a31.head()
```

```python
# grouping by Match Id to get runs scored in a match and player name to get runs scored by players  
a32 = a31.groupby(['Match_Id','Player_Name'])['Batsman_Scored'].sum().sort_values(ascending=False)
```

```python
# selecting players who scored hundred or more
a33 = a32[a32>99]
pd.DataFrame(a33)
```

```python
# counting the number of centuries
pd.DataFrame(a33.groupby('Player_Name').count().sort_values(ascending=False).head(10))
```

# Q4 Top 5 players with most wickets? 

```python
# merging files ball_by_ball and player to get bowler names on bowler_id
a41 = pd.merge(ball_by_ball,player, left_on= 'Bowler_Id', right_on = 'Player_Id',how='inner') 
a41.head(1)
```

```python
# Checking types of dissimals
a41['Dissimal_Type'].unique()
```

```python
# Since runout , obstruction of the field and retired hurt doesn't count as bowler wicket therefore removing these dissmals
a42 = a41[a41['Dissimal_Type'].isin(['caught','caught and bowled','lbw','bowled','hit wicket','stumped'])]
a42.head()
```

```python
# selecting top 5 bowlers with wickets
a43 = a42.groupby('Player_Name')['Dissimal_Type'].count().sort_values(ascending=False)
pd.DataFrame(a43.head())
```

# Q5 Bowlers runs conceded in every match

```python
# Some extras(legbyes & byes) are not included in runs conceded by bowlers 
a41['Extra_Type'].unique()
```

```python
# Calculating extra runs given by bowlers(leg byes and byes are not considered as runs conceded by bowlers)
a51 = a41[a41['Extra_Type'].isin(['wides', 'noballs', 'penalty'])]
a52 = a51.groupby(['Match_Id','Player_Name'])[['Extra_Runs']].sum()
```

```python
a52.head()
```

```python
# Calculating runs given by bowlers(scored by batsman)
a53 = a41.groupby(['Match_Id','Player_Name'])[['Batsman_Scored']].sum()
a53.head()
```

```python
# merging extra runs and runs scored by batsman in a dataframe
a54 = pd.merge(a52,a53,on =['Match_Id','Player_Name'],how='outer')
a54.head()
```

```python
# Checking for null values as some overs were maiden or some bowlers does not gave any extras
a54.isnull().sum() 
```

```python
# Since bowlers does not conceded any runs or given extra filling null values with 0
a55 = a54.fillna(0)
```

```python
a55.isnull().sum()
```

```python
# Creating a new column which is sum of Extra Runs and Batsman scored 
a55['Runs_Conceded'] = a55.Extra_Runs + a55.Batsman_Scored
a55.sort_values(by='Runs_Conceded')
a55
```
